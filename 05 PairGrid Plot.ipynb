```python
import  pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('darkgrid')
```

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
x=sns.PairGrid(iris)
x=x.map(plt.scatter)
```

```python
x=sns.PairGrid(iris)
x=x.map_diag(plt.hist)
```

```python
x=sns.PairGrid(iris)
x=x.map_diag(plt.hist)
x=x.map_offdiag(plt.scatter)
```

```python
x=sns.PairGrid(iris,hue='species',palette='husl')
x=x.map_diag(plt.hist)
x=x.map_upper(plt.scatter)
x=x.map_lower(sns.kdeplot)
x=x.add_legend()
```

```python
x=sns.PairGrid(iris,hue='species',palette='husl')
x=x.map_diag(plt.hist,histtype='step',linewidth=3)
x=x.map_upper(plt.scatter)
x=x.map_lower(plt.scatter)
x=x.add_legend()
```

```python
x=sns.PairGrid(iris,vars=['petal_length','petal_width'])
x=x.map(plt.scatter)
```

```python
x=sns.PairGrid(iris,x_vars=['petal_length','petal_width'],
              y_vars=['sepal_length','sepal_width'])
x=x.map(plt.scatter)
```

```python
x=sns.PairGrid(iris,hue='species',vars=['petal_length','petal_width'])
x=x.map_diag(plt.hist,edgecolor='red')
x=x.map_offdiag(plt.scatter,edgecolor='black')
x=x.add_legend()
```

```python
x=sns.PairGrid(iris,hue='species',hue_kws={'marker':['D','s','+']})
x=x.map(plt.scatter,s=40,edgecolor='black',alpha=0.5)
x=x.add_legend()
```
