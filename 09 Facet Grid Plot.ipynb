```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('darkgrid')
```

```python
tips=sns.load_dataset('tips')
```

```python
sns.FacetGrid(row='smoker',col='time',data=tips)
```

```python
x=sns.FacetGrid(tips,row='smoker',col='time')
x=x.map(plt.hist,'total_bill')
```

```python
x=sns.FacetGrid(tips,row='smoker',col='time')
x=x.map(plt.hist,'total_bill',color='green',bins=20)
```

```python
x=sns.FacetGrid(tips,row='time',col='smoker')
x=x.map(plt.scatter,'total_bill','tip')
```

```python
x=sns.FacetGrid(tips,row='time',col='smoker')
x=x.map(sns.regplot,'total_bill','tip')
```

```python
x=sns.FacetGrid(tips,row='time',col='smoker',hue='sex')
x=(x.map(plt.scatter,'total_bill','tip').add_legend())
```

```python
x=sns.FacetGrid(tips,row='sex',col='day')
x=x.map(sns.boxplot,'total_bill','time')
```

```python
x=sns.FacetGrid(tips,col='day',size=5,aspect=0.5,col_order=['Sun','Sat','Fri','Thur'])
x=x.map(sns.boxplot,'time','total_bill',color='red')
```

```python
x=sns.FacetGrid(tips,row='time',col='smoker',hue='sex',palette='winter_r')
x=x.map(plt.scatter,'total_bill','tip')
```
