```python
import  pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('darkgrid')
```

```python
fl=sns.load_dataset('flights')
```

```python
fl.head()
```

```python
fl=fl.pivot('month','year','passengers')
```

```python
sns.heatmap(fl)
```

```python
sns.clustermap(fl)
```

```python
sns.clustermap(fl,col_cluster=False)
```

```python
sns.clustermap(fl,row_cluster=False)
```

```python
sns.clustermap(fl,cmap='viridis_r')
```

```python
sns.clustermap(fl,cmap='viridis_r',lw=0.5)
```

```python
sns.clustermap(fl,cmap='Blues_r',lw=0.5)
```

```python
sns.clustermap(fl,cmap='summer_r',lw=0.5,figsize=(15,6))
```

```python
# standerdize across col or rows
```

```python
sns.clustermap(fl,standard_scale=1)
```

```python
sns.clustermap(fl,standard_scale=0)
```

```python
# normalizing our dataset = z_score 0/1
```

```python
sns.clustermap(fl, z_score=1)
```

```python
sns.clustermap(fl, z_score=0,)
```
