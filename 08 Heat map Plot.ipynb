```python
import  pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('darkgrid')
```

```python
nor=np.random.rand(9,6)
```

```python
sns.heatmap(nor)
```

```python
sns.heatmap(nor,annot=True)
```

```python
sns.heatmap(nor,annot=True,vmin=0,vmax=2)
```

```python
fl=sns.load_dataset('flights')
```

```python
fl.head()
```

```python
fl=fl.pivot('month','year','passengers')
sns.heatmap(fl)
```

```python
sns.heatmap(fl,annot=True,fmt='d')
```

```python
sns.heatmap(fl,lw=0.5)
```

```python
sns.heatmap(fl,annot=True,fmt='d',lw=0.5,vmin=100,vmax=700)
```

```python
sns.heatmap(fl,center=fl.loc['May',1960],annot=True,fmt='d')
```

```python
sns.heatmap(fl,cmap='summer',cbar=False,lw=0.5)
```
