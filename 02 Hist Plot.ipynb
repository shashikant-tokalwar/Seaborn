```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('darkgrid')
```

```python
num=np.random.randn(150)
```

```python
sns.distplot(num)
```

```python
sns.distplot(num,color='red',axlabel='Data',label='Shashi')
plt.legend()
plt.show()
```

```python
label_dist=pd.Series(num,name='variable x')
```

```python
sns.distplot(label_dist)
```

```python
sns.distplot(label_dist,vertical=True)
```

```python
sns.distplot(label_dist,vertical=True,color='red')
```

```python
sns.distplot(label_dist,hist=False)
```

```python
sns.distplot(label_dist,rug=True,hist=False,color='red')
```

```python
sns.distplot(label_dist,bins=30)
```
