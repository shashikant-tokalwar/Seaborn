```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('darkgrid')
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.regplot(x='total_bill',y='tip',data=tips)
```

```python
sns.regplot(x='total_bill',y='tip',data=tips,color='purple')
```

```python
sns.regplot(x='total_bill',y='tip',data=tips,color='red',
           marker='D')
```

```python
sns.regplot(x='total_bill',y='tip',data=tips,marker='o',
           scatter_kws={'color':'green'},
           line_kws={'color':'red','lw':0.9})
```

```python
sns.regplot(x='total_bill',y='tip',data=tips,marker='D',
           scatter_kws={'color':'purple'},
           line_kws={'color':'black','lw':0.9})
```

```python
# To control Error Bar--ci
sns.regplot(x='total_bill',y='tip',data=tips,marker='D',
           scatter_kws={'color':'olive'},
           line_kws={'color':'black','lw':0.9},ci=1)
```

```python
sns.regplot(x='total_bill',y='tip',data=tips,marker='D',
           scatter_kws={'color':'olive'},
           line_kws={'color':'black','lw':0.9},ci=100)
```

```python
sns.regplot(x='size',y='total_bill',data=tips)
```

```python
sns.regplot(x='size',y='total_bill',data=tips,x_jitter=0.1,
           scatter_kws={'color':'red'})
```

```python
sns.regplot(x='size',y='total_bill',data=tips,
           x_estimator=np.mean)
```
