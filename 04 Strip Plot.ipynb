```python
import  pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('darkgrid')
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.stripplot(x=tips['tip'])
```

```python
sns.stripplot(x=tips['total_bill'],color='green')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=False)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=0.3)
```

```python
sns.stripplot(x='total_bill',y='day',data=tips)
```

```python
sns.stripplot(x='total_bill',y='day',data=tips,jitter=0.3,
             linewidth=1.2)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',dodge=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='smoker',
              dodge=True,palette='winter_r')
```

```python
sns.stripplot(x='sex',y='tip',data=tips,palette='husl')
```

```python
sns.stripplot(x='sex',y='tip',data=tips,palette='husl',marker='D')
```

```python
sns.stripplot(x='sex',y='total_bill',data=tips,marker='D',size=10,
             edgecolor='black',alpha=0.2,linewidth=1.5)
```

```python
sns.stripplot(x='tip',y='day',data=tips,palette='autumn')
sns.boxplot(x='tip',y='day',data=tips,palette='winter_r')
```

```python
sns.stripplot(x='tip',y='day',data=tips)
sns.violinplot(x='tip',y='day',data=tips,palette='coolwarm')
```
