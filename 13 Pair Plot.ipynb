```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('darkgrid')
```

```python
iris=sns.load_dataset('iris')
tips=sns.load_dataset('tips')
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue='species')
```

```python
sns.pairplot(iris,hue='species',palette='husl')
```

```python
sns.pairplot(iris,hue='species',markers=['o','D','+'],palette='winter_r')
```

```python
# to see limited plot pass argument [vars]
sns.pairplot(iris,vars=['sepal_length','sepal_width'])
```

```python
sns.pairplot(tips,vars=['total_bill','tip'])
```

```python
sns.pairplot(tips,vars=['total_bill','tip'],size=5)
```

```python
# x_vars, y_vars
```

```python
sns.pairplot(iris, x_vars=['petal_length','petal_width'],
            y_vars=['sepal_length','sepal_width'])
```

```python
sns.pairplot(iris, x_vars=['petal_length','petal_width'],
            y_vars=['sepal_length','sepal_width'],palette='husl',hue='species')
```

```python
sns.pairplot(iris, x_vars=['petal_length'],
            y_vars=['sepal_width'])
```

```python
# diag_kind
```

```python
sns.pairplot(iris, diag_kind='kde')
```

```python
sns.pairplot(iris, diag_kind='kde',hue='species')
```

```python
# To add reg line 
sns.pairplot(iris,kind='reg',hue='species',palette='husl')
```

```python
sns.pairplot(tips,kind='reg',hue='sex')
```
